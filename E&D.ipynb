```python
from ipywidgets import interact, widgets

def caesar_cipher(message, shift, choice):
    result = ""

    # Change shift value for decryption
    if choice.lower() == 'decrypt':
        shift = -shift

    # Go through each character in the message
    for char in message:
        if char.isalpha():
            if char.isupper():
                new_char = chr((ord(char) - 65 + shift) % 26 + 65)
                result += new_char
            elif char.islower():
                new_char = chr((ord(char) - 97 + shift) % 26 + 97)
                result += new_char
        else:
            result += char

    return result

# --- Interactive Interface ---
def run_cipher(message, shift, choice):
    output = caesar_cipher(message, shift, choice)
    print("üîê Result:", output)

print("=== Caesar Cipher Encryption and Decryption ===")

# Interactive widgets for Colab
interact(
    run_cipher,
    message=widgets.Text(value="Hello World", description="Message:"),
    shift=widgets.IntSlider(value=3, min=1, max=25, description="Shift:"),
    choice=widgets.Dropdown(options=['encrypt', 'decrypt'], description="Mode:")
)
```
